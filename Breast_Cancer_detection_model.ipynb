```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_acc = []
test_acc = []

for i in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    train_acc.append(knn.score(x_train, y_train))
    test_acc.append(knn.score(x_test, y_test))

plt.plot(range(1, 20), train_acc, label="Train Accuracy")
plt.plot(range(1, 20), test_acc, label="Test Accuracy")
plt.xticks(np.arange(1, 20, 1))
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

We selected k = 5 because it offers stable generalization with a small trainâ€“test gap, avoiding both overfitting at low k and underfitting at high k.

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
y_pred_knn=knn.predict(x_test)
```

```python
from sklearn.svm import SVC
```

```python
kernel=['linear','rbf','poly','sigmoid']
for i in kernel:
  sv1=SVC(kernel=i)
  print(f"for kernel={i}")
  sv1.fit(x_train,y_train)
  print(sv1.score(x_train,y_train))
  print(sv1.score(x_test,y_test))
```

Among the evaluated SVM kernels, the linear kernel was selected as it achieved high and stable performance on both training and testing data with minimal generalization gap. This indicates that the breast cancer dataset is largely linearly separable, and using a linear kernel avoids unnecessary model complexity while maintaining strong predictive performance.

```python
sv=SVC(kernel='linear')
sv.fit(x_train,y_train)
```

```python
y_pred_svm=sv.predict(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='entropy',max_depth=3)
dt.fit(x_train,y_train)
```

```python
y_pred_dt=dt.predict(x_test)
```

The maximum depth of the decision tree was selected by analyzing training and testing accuracy across different depths. At depth = 3, the model achieved high test accuracy with a negligible difference between training and testing performance. This indicates an optimal balance between bias and variance, and therefore depth = 3 was selected to avoid overfitting.

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(8,8))
plot_tree(dt)
plt.show()
```

The decision tree analysis reveals that feature x[27] is the most influential predictor of breast cancer diagnosis, as it appears at the root node with the highest information gain. Early splits result in a significant reduction in entropy, demonstrating strong class separability. Several leaf nodes achieve zero entropy, indicating confident and pure classifications. Secondary tumor characteristics further refine predictions, highlighting the importance of multiple interacting features. The controlled depth of the tree prevents overfitting while maintaining interpretability, making the model both accurate and clinically explainable.

```python
from sklearn.linear_model import LogisticRegression
```

```python
C_values = [0.001, 0.01, 0.1, 1, 10, 100]
for c in C_values:
    lr = LogisticRegression(C=c, max_iter=1000)
    lr.fit(x_train, y_train)
    print(
        "C =", c,
        "Train =", lr.score(x_train,y_train),
        "Test =", lr.score(x_test,y_test)
    )
```

```python
lr_final = LogisticRegression(C=0.1, max_iter=1000)
lr_final.fit(x_train,y_train)
print("Final Train Accuracy:", lr_final.score(x_train, y_train))
print("Final Test Accuracy:", lr_final.score(x_test, y_test))
```

```python
y_pred_lr=lr_final.predict(x_test)
```

Logistic Regression was implemented as a baseline classification model. Feature scaling was applied prior to training to ensure numerical stability. The regularization parameter C was tuned to control model complexity. Smaller C values resulted in underfitting, while larger values showed signs of overfitting. An intermediate C value achieved the highest test accuracy with minimal difference between training and testing performance, indicating good generalization. The final model demonstrated strong recall for malignant cases, making it suitable for medical diagnosis.

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,confusion_matrix
from sklearn.metrics import precision_score,recall_score,f1_score
```

```python
cm_knn=confusion_matrix(y_test,y_pred_knn)
cm_svm=confusion_matrix(y_test,y_pred_svm)
cm_dt=confusion_matrix(y_test,y_pred_dt)
cm_lr=confusion_matrix(y_test,y_pred_lr)
```

```python
confusion_matrix_list = [cm_knn, cm_svm, cm_dt, cm_lr]
for i in confusion_matrix_list:
    sns.heatmap(i, annot=True)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
y_pred_list = [y_pred_knn, y_pred_svm, y_pred_dt, y_pred_lr]
model_list = ['KNN', 'SVM', 'Decision Tree', 'Logistic Regression']
for model, y_pred in zip(model_list, y_pred_list):
    print(f'\nModel: {model}')
    print('Accuracy :', accuracy_score(y_test, y_pred))
    print('Precision:', precision_score(y_test, y_pred))
    print('Recall   :', recall_score(y_test, y_pred))
    print('F1-score :', f1_score(y_test, y_pred))

```

Interpretation of False Positive and False Negative Cases

In breast cancer diagnosis, prediction errors have different clinical consequences. A false positive (FP) occurs when a benign tumor is incorrectly classified as malignant. Although this may cause anxiety and lead to additional diagnostic tests such as biopsies, it does not pose an immediate life-threatening risk. In contrast, a false negative (FN) occurs when a malignant tumor is incorrectly classified as benign. This is far more dangerous because it may delay diagnosis and treatment, allowing the cancer to progress to an advanced stage. Therefore, minimizing false negative cases is a critical requirement in breast cancer detection systems.

Interpretation of Precision, Recall, and F1-Score

Precision measures the proportion of correctly predicted malignant cases among all cases predicted as malignant. High precision indicates fewer false positive cases and ensures that when the model predicts cancer, the prediction is reliable.

Recall (Sensitivity) measures the proportion of actual malignant cases that are correctly identified by the model. High recall is essential in cancer diagnosis, as it minimizes false negatives and ensures that most cancer cases are detected.

F1-Score represents the harmonic mean of precision and recall. It provides a balanced measure of model performance, particularly when there is a trade-off between detecting cancer cases and avoiding false alarms.

In medical diagnosis, recall is often prioritized over precision because missing a cancer case has more severe consequences than incorrectly flagging a benign case.

Model Comparison Based on Medical Priorities

KNN achieved perfect precision but had the lowest recall, indicating that although its cancer predictions were accurate, it missed several malignant cases, which is undesirable in medical diagnosis.

Decision Tree achieved the highest recall, meaning it detected most malignant cases. However, it produced more false positives and showed lower overall accuracy, increasing the likelihood of unnecessary medical interventions.

Logistic Regression demonstrated balanced performance with good precision and recall, along with strong interpretability, making it a reliable baseline model.

Support Vector Machine (Linear Kernel) achieved the highest accuracy and F1-score while maintaining strong recall and precision, indicating excellent generalization and reliable cancer detection.

Final Conclusion and Best Model Selection

Considering the clinical importance of minimizing false negatives while maintaining overall stability, the Support Vector Machine (Linear Kernel) was selected as the best model for breast cancer diagnosis. It provides a strong balance between high recall and precision, ensuring that most malignant tumors are detected without excessive false alarms. While the Decision Tree model exhibited slightly higher recall, its lower precision and risk of overfitting reduce its reliability. Therefore, the SVM model offers the most effective and medically suitable solution for breast cancer diagnostic classification.
